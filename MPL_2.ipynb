```python
import matplotlib.pyplot as plt

carDeets = {}
noOfFeatures = int(input("Enter no. of features: "))

for i in range(noOfFeatures):
    featureName = input(f"Enter feature {i+1} name: ")           # ✅ Fixed
    featureVals = input("Enter values (space-separated): ")
    listOfFeatVals = [float(val) for val in featureVals.split()]  # ✅ List of numbers
    carDeets[featureName] = listOfFeatVals                        # ✅ Proper assignment

print("Car data:", carDeets)

lineCol = input("Enter line colour (e.g., blue): ")
markerStyle = input("Enter marker style (e.g., o, s, ^): ")
markerCol = input("Enter marker colour (e.g., red): ")

plotStyle = {
    'markerfacecolor': markerCol,    # ✅ Fixed param
    'marker': markerStyle,
    'color': lineCol,
    'linewidth': 2,
    'markersize': 8
}

plt.figure(figsize=(12, 8))

# Plot each feature's data points
for featureName, values in carDeets.items():
    x_indices = list(range(len(values)))  # [0,1,2,3...] for x-axis
    plt.plot(x_indices, values,           # ✅ Proper x,y unpacking
             label=featureName, 
             **plotStyle)                 # ✅ Unpack style dict

plt.title("Car Features Comparison")
plt.xlabel("Data Point Index")
plt.ylabel("Feature Values")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
