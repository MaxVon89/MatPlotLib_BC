```python
import matplotlib.pyplot as plt
import random
```

```python
x,y = [],[]

for i in range(1,11):
    x.append(i)
    y.append(random.randint(1,100))

plt.plot(x,y)
plt.show()
```

```python
import numpy as np

x2 = np.array([])
y2 = np.array([])
y2_1 = np.array([])

for i in range(1,100):
    x2 = np.append(x2, i)
    y2 = np.append(y2, random.randint(1,100))
    y2_1 = np.append(y2_1, random.randint(1,80))

# plt.plot(x2,[y2,y2_1])
plt.plot(x2,y2)
plt.plot(x2,y2_1)
plt.show()
```
