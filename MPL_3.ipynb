## Labelling

```python
import matplotlib.pyplot as plt
import random as r
import numpy as np

x1 = np.array([i for i in range(10)])
y1 = np.array([r.randint(1, 100) for _ in range(10)])

plt.title("Randomness", fontsize=10, fontweight=20, family="Garamond")
plt.xlabel("Progression", fontsize=7, family="Times New Roman")
plt.ylabel("Values", fontsize=7, family="Times New Roman")

plt.xticks(x1)
# plt.yticks(y1)

plt.plot(x1,y1)
plt.show()
```
